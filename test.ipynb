```python
%%writefile tests/test_mutant.py
from mutant import is_mutant


def test_enunciado_mutante():
    dna = ["ATGCGA", "CAGTGC", "TTATGT", "AGAAGG", "CCCCTA", "TCACTG"]
    assert is_mutant(dna) is True


def test_humano_basico():
    dna = ["ATGCGA", "CAGTGC", "TTATTT", "AGACGG", "GCGTCA", "TCACTG"]
    assert is_mutant(dna) is False


def test_mutante_horizontal():
    dna = ["AAAAAA", "CAGTGC", "TTATGT", "AGAAGG", "CCCCTA", "TCACTG"]
    assert is_mutant(dna) is True


def test_mutante_vertical():
    dna = [
        "ATGCGA",
        "ATGTGC",
        "ATATGT",
        "ATAAGG",
        "ACCCTA",
        "TCACTG",
    ]
    assert is_mutant(dna) is True


def test_mutante_diagonal_principal():
    dna = [
        "ATGCGA",
        "CAGAGC",
        "TTATGT",
        "AGAAAG",
        "CCCCTA",
        "TCACTG",
    ]
    assert is_mutant(dna) is True


def test_mutante_diagonal_inversa():
    dna = [
        "ATGCAA",
        "CAGTAC",
        "TAATGT",
        "AGAAGG",
        "CCCCTA",
        "TCACTG",
    ]
    assert is_mutant(dna) is True


def test_matriz_vacia_es_humano():
    assert is_mutant([]) is False


def test_caracteres_invalidos_no_crashea():
    dna = [
        "ATGCXA",
        "CAGTGC",
        "TTATGT",
        "AGAAGG",
        "CCCCTA",
        "TCACTG",
    ]
    assert is_mutant(dna) in (True, False)
```
